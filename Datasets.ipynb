```python
# Import and Load the Dataset
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data1 = pd.read_csv('clinical_data_breast_cancer.csv')
```

#### Question 1
#### How many patients are male and how many are female?

```python
# Count the number of male and female patients
gender_count = data1['Gender'].value_counts()

print(gender_count)

# Create a bar plot
plt.bar(gender_count.index, gender_count.values)
plt.title('Gender Distribution in Breast Cancer Clinical Data')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

#### Question 2
#### What is the age range of patients included in the dataset?

```python
# Finding the minimum and maximum age values
min_age = data1['Age at Initial Pathologic Diagnosis'].min()
max_age = data1['Age at Initial Pathologic Diagnosis'].max()

# Printing the age range of patients included in the dataset
print(f"The age range of patients included in the dataset is from {min_age} to {max_age} years old.")
```

```python
# Calculating the frequency of every age
age_freq = data1['Age at Initial Pathologic Diagnosis'].value_counts()

# Sorting the age frequencies in ascending order
age_freq = age_freq.sort_index()

# Creating a bar graph of age frequencies
plt.bar(age_freq.index, age_freq.values)

# Adding labels and title to the graph
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Breast Cancer Patients')

# Displaying the graph
plt.show()
```

#### Question 3
#### What is the distribution of breast cancer stages (I-IV) in the dataset?

```python
# Extracting breast cancer stage information
breast_cancer_stages = data1['Tumor'].value_counts()

# Creating a pie chart of breast cancer stage distribution
plt.pie(breast_cancer_stages.values, labels=breast_cancer_stages.index, autopct='%1.1f%%')

# Adding title to the graph
plt.title('Distribution of Breast Cancer Stages in Dataset')

# Displaying the graph
plt.show()
```

#### Question 4
#### Is there a correlation between the patient's age and tumor size?

```python
# Create a scatter plot of age and tumor size
plt.scatter(data1['Age at Initial Pathologic Diagnosis'], data1['Tumor'])
plt.title('Patient Age vs Tumor Size')
plt.xlabel('Age at Initial Pathologic Diagnosis')
plt.ylabel('Tumor Size')
plt.show()
```

#### Question 5
#### How many patients have positive ER status, PR status, and HER2 status?

```python
# Count the number of patients with positive ER status
er_status_count = data1['ER Status'].value_counts()['Positive']
print(f"Number of patients with positive ER status: {er_status_count}")

# Count the number of patients with positive PR status
pr_status_count = data1['PR Status'].value_counts()['Positive']
print(f"Number of patients with positive PR status: {pr_status_count}")

# Count the number of patients with positive HER2 status
her2_status_count = data1['HER2 Final Status'].value_counts()['Positive']
print(f"Number of patients with positive HER2 status: {her2_status_count}")

# Create a bar plot
counts = [er_status_count, pr_status_count, her2_status_count]
labels = ['ER Status', 'PR Status', 'HER2 Final Status']
plt.bar(labels, counts)
plt.title('Number of Patients with Positive ER, PR, and HER2 Status')
plt.xlabel('Status')
plt.ylabel('Count')
plt.show()
```

#### Question 6
#### What is the distribution of HER2 status in the dataset, and how many samples are HER2-positive and HER2-negative?

```python
# Count the number of HER2-positive and HER2-negative samples
her2_counts = data1['HER2 Final Status'].value_counts()

# Print the counts
print('HER2-positive samples:', her2_counts['Positive'])
print('HER2-negative samples:', her2_counts['Negative'])

# Create a bar chart of the HER2 status counts
fig, ax = plt.subplots()
her2_counts.plot(kind='bar', ax=ax)

# Set the chart title and axis labels
ax.set_title('HER2 Status Distribution')
ax.set_xlabel('HER2 Status')
ax.set_ylabel('Number of Samples')

# Add data labels to the bars
for i, v in enumerate(her2_counts):
    ax.text(i, v+10, str(v), ha='center')

# Display the chart
plt.show()
```

#### Question 7
#### What is the average number of lymph nodes involved?

```python
# count the number of positive and negative nodes in patients
node_counts = data1['Node-Coded'].value_counts()
print("Total nodes:", node_counts)

# create a bar chart
plt.bar(node_counts.index, node_counts.values)

# add labels and title to the chart
plt.xlabel('Node Status')
plt.ylabel('Number of Nodes')
plt.title('Total Number of Positive and Negative Nodes')
plt.show()
```

#### Question 8
#### What is the correlation between the age at initial diagnosis and the number of positive lymph nodes?

```python
# Create a scatter plot of Age at Initial Pathologic Diagnosis vs. Node-Coded
plt.scatter(data1['Age at Initial Pathologic Diagnosis'], data1['Node-Coded'])
plt.xlabel('Age at Initial Pathologic Diagnosis')
plt.ylabel('Number of Positive Nodes')
plt.title('Correlation between Age and Number of Positive Nodes')
plt.show()
```

#### Question 9
#### What is the prevalence of estrogen receptor (ER) positive and progesterone receptor (PR) positive breast cancers in the dataset?

```python
# Count the number of ER and PR positive cases
er_pos_count = (data1['ER Status'] == 'Positive').sum()
pr_pos_count = (data1['PR Status'] == 'Positive').sum()

# Calculate the prevalence of ER and PR positive cases
er_pos_prevalence = er_pos_count / len(data1) * 100
pr_pos_prevalence = pr_pos_count / len(data1) * 100

# Print the results
print(f"ER positive prevalence: {er_pos_prevalence:.2f}%")
print(f"PR positive prevalence: {pr_pos_prevalence:.2f}%")

# Create a bar plot
labels = ['ER Positive', 'PR Positive']
values = [er_pos_count, pr_pos_count]
plt.bar(labels, values)
plt.title('Prevalence of ER and PR Positive Breast Cancers')
plt.xlabel('Receptor Status')
plt.ylabel('Number of Cases')
plt.show()
```

#### Question 10
#### What are the median values for each breast cnacer cluster?

```python
# Select the columns of interest
cols_of_interest = ['miRNA Clusters', 'methylation Clusters', 
                    'CN Clusters', 'Integrated Clusters (with PAM50)', 
                    'Integrated Clusters (no exp)', 'Integrated Clusters (unsup exp)']

# Calculate the medians of the columns
medians = data1[cols_of_interest].median(numeric_only=True)

# Print the medians
print(medians)
```

#### Question 11
#### How many patients have each of the PAM50 mRNA subtypes

```python
# Group the patients by PAM50 subtype and count the number of patients in each group
subtype_counts = data1.groupby('PAM50 mRNA').size()

# Print the counts for each subtype
print(subtype_counts)

# Create a bar plot
plt.bar(subtype_counts.index, subtype_counts.values)

# Set the axis labels and title
plt.xlabel('PAM50 mRNA Subtype')
plt.ylabel('Number of Patients')
plt.title('Number of Patients in Each PAM50 mRNA Subtype')

# Show the plot
plt.show()
```

#### Question 12
#### How many patients have each of the integrated clusters (with PAM50)?

```python
# Count the number of patients in each cluster
counts = data1['Integrated Clusters (with PAM50)'].value_counts()
print(counts)

# Create a bar chart
fig, ax = plt.subplots(figsize=(8,6))
data1.plot(kind='bar', ax=ax)

# Set plot title and labels
ax.set_title('Number of Patients in Each Integrated Cluster', fontsize=16)
ax.set_xlabel('Integrated Clusters (with PAM50)', fontsize=14)
ax.set_ylabel('Number of Patients', fontsize=14)

# Display the plot
plt.show()
```
